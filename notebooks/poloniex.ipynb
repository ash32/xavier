```python
dates = pd.date_range(start=1404172800e9, end=datetime.utcnow(), freq='30min')
# dates = pd.date_range(start=1404172800e9, end=1404382800e9, freq='30min')
dates = pd.DataFrame(dates).set_index(0)
```

```python
def get_coin_symbol(filename):
    return filename[4:-5]

def create_data_frame(directory, filename):
    cols = ['close', 'high', 'low']
    symbol = get_coin_symbol(filename)
    
    df = pd.read_json(os.path.join(directory, filename))
    df = df.set_index('date')
    df = df[cols]
    df = df.rename(columns={c: c+'_'+symbol for c in cols})
    return df
```

```python
import functools
files = os.listdir('data/raw/poloniex')
# files = ['BTC_ETH.json']
dfs = [create_data_frame('data/raw/poloniex', f) for f in files]
bigdf = functools.reduce(lambda x, y: x.join(y), dfs, dates)
bigdf = bigdf.fillna(method='bfill')
bigdf = bigdf.dropna(how='all')
bigdf
```
