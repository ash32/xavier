### Price relative vector (y_t)

Most basic strategy. Choose a distribution (i.e. all 33%) and rebalance every period to maintain that distribution.

```python
# Create price relative vector
closing_df = rel.test_df.iloc[:, 0::3]  # get all closing prices
closing_df
```

```python
close_rel = closing_df.values[1:]/closing_df.values[:-1]
cash_col = np.ones((close_rel.shape[0], 1))
price_rel_vec = np.concatenate((cash_col, close_rel), axis=1)
price_rel_vec
```

```python
def normalize(v, axis=None):
    norm = np.linalg.norm(v, ord=1, axis=axis)
    if axis:
        norm = np.expand_dims(norm, axis=axis)
    return v/norm
```

```python
def const_port_dist(closing_price_vec, prev_weight_vec):
    return np.ones(closing_price_vec.size) / closing_price_vec.size
```

```python
def get_portfolio_weights(price_rel_vec, dist_func, initial_weight_vec):
    weights = np.zeros(price_rel_vec.shape)
    weights[0] = normalize(initial_weight_vec)
    
    for i in range(1, price_rel_vec.shape[0]):
        weights[i] = normalize(dist_func(price_rel_vec[i-1], weights[i-1]))
        
    return weights
    
def get_portfolio_values(price_rel_vec, weights):
    portf_val = np.sum(price_rel_vec * weights, axis=1)
    return np.cumprod(portf_val)
```

```python
initial_weights = np.ones(price_rel_vec.shape[1]) / price_rel_vec.shape[1]
weights = get_portfolio_weights(price_rel_vec, const_port_dist, initial_weights)
weights
```

```python
portf_val = get_portfolio_values(price_rel_vec, _)
portf_val
```

```python
def plot_portfolio_values(portf_val):
    x = range(portf_val.size)
    plt.plot(portf_val)
    plt.show()
    
plot_portfolio_values(portf_val)
```

## BUY AND HOLD

Here, we want to have every portfolio distribution just be the same as the previous one dotted with the closing price of the last period. This will simulate buying the currency and not deviating from it after.

```python
def buy_and_hold_dist(closing_price_vec, prev_weight_vec):
    return normalize(closing_price_vec * prev_weight_vec)
```

```python
initial_weights = np.array([1,2,1])
bh_weights = get_portfolio_weights(price_rel_vec, buy_and_hold_dist, initial_weights)
bh_weights
```

```python
bh_portf_val = get_portfolio_values(price_rel_vec, bh_weights)
bh_portf_val
```

```python
plot_portfolio_values(bh_portf_val)
```
